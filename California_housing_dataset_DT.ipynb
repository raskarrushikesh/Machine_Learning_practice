**Use California housing dataset to build a model with DT to predict median_house_value**

```python
import numpy as np
import pandas as pd
```

```python
#Load the dataset for training
train=pd.read_csv('/content/sample_data/california_housing_train.csv')
train.head()
```

```python
#Load the dataset for testing
test=pd.read_csv('/content/sample_data/california_housing_test.csv')
test.head()
```

```python
train.shape
```

```python
test.shape
```

```python
X_train=train.drop(columns='median_house_value')
y_train=train['median_house_value']
X_test=test.drop(columns='median_house_value')
y_test=test['median_house_value']
print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(X_train,y_train)
```

```python
DecisionTreeRegressor?
```

```python
y_pred=model.predict(X_test)
```

```python
model.criterion # by default
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print("R2_Score: ",r2_score(y_test,y_pred))
print("MAE: ",mean_absolute_error(y_test,y_pred))
print("MSE :",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
model_p=DecisionTreeRegressor(criterion='poisson')
model_p.fit(X_train,y_train)
y_pred_p=model_p.predict(X_test)
print("R2_Score: ",r2_score(y_test,y_pred_p))
print("MAE: ",mean_absolute_error(y_test,y_pred_p))
print("MSE :",mean_squared_error(y_test,y_pred_p))
```

```python
from sklearn.tree import DecisionTreeRegressor
model_f=DecisionTreeRegressor(criterion='friedman_mse')
model_f.fit(X_train,y_train)
y_pred_f=model_f.predict(X_test)
print("R2_Score: ",r2_score(y_test,y_pred_f))
print("MAE: ",mean_absolute_error(y_test,y_pred_f))
print("MSE :",mean_squared_error(y_test,y_pred_f))
```

```python
from sklearn.tree import DecisionTreeRegressor
model_a=DecisionTreeRegressor(criterion='absolute_error')
model_a.fit(X_train,y_train)
y_pred_a=model_a.predict(X_test)
print("R2_Score: ",r2_score(y_test,y_pred_a))
print("MAE: ",mean_absolute_error(y_test,y_pred_a))
print("MSE :",mean_squared_error(y_test,y_pred_a))
```
